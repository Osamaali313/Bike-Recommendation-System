```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Replace 'bikes_data.csv' with the actual file path if it's located in a different directory
file_path = 'bikes_data.csv'

# Read the CSV file into a pandas DataFrame
data_df = pd.read_csv(file_path)

# Convert the 'Torque' column to a numeric data type
data_df['Torque'] = pd.to_numeric(data_df['Torque'], errors='coerce')

# Check the data type of the 'Transmission Type' column
# If it is an object (string) data type, convert it to a categorical data type
if data_df['Transmission Type'].dtype == 'object':
    data_df['Transmission Type'] = data_df['Transmission Type'].astype('category')

# Get the top 5 most frequent transmission types
top_5_transmission_types = data_df['Transmission Type'].value_counts().nlargest(5).index

# Filter the DataFrame to include only the top 5 transmission types
data_top_5 = data_df[data_df['Transmission Type'].isin(top_5_transmission_types)]

# Create a violin plot to compare torque by the top 5 transmission types
plt.figure(figsize=(10, 6))
sns.violinplot(x='Transmission Type', y='Torque', data=data_top_5)
plt.xlabel('Transmission Type')
plt.ylabel('Torque')
plt.title('Comparison of Torque by Top 5 Transmission Types')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Replace 'bikes_data.csv' with the actual file path if it's located in a different directory
file_path = 'bikes_data.csv'

# Read the CSV file into a pandas DataFrame
data_df = pd.read_csv(file_path)

# Data Cleaning Steps

# 1. Convert 'Horsepower' and 'Number of Seating' to numeric data types
data_df['Horsepower'] = pd.to_numeric(data_df['Horsepower'], errors='coerce')
data_df['Number of Seating'] = pd.to_numeric(data_df['Number of Seating'], errors='coerce')

# 2. Convert 'Price (in INR)' to numeric data type
data_df['Price (in INR)'] = pd.to_numeric(data_df['Price (in INR)'], errors='coerce')

# 3. Drop rows with missing values in 'Horsepower', 'Number of Seating', and 'Price (in INR)' columns
data_df.dropna(subset=['Horsepower', 'Number of Seating', 'Price (in INR)'], inplace=True)

# Feature Engineering Steps

# 1. Create a new feature for 'Price Category'
# Group prices into categories based on some logical intervals
price_bins = [0, 500000, 1000000, 1500000, np.inf]
price_labels = ['Low', 'Medium', 'High', 'Very High']
data_df['Price Category'] = pd.cut(data_df['Price (in INR)'], bins=price_bins, labels=price_labels)

# 2. Extract the 'Brand' from the 'Model' column
# Assuming the 'Model' column follows a format like "Brand ModelName"
data_df['Brand'] = data_df['Model'].apply(lambda x: x.split()[0].lower())

# 3. Calculate the 'Power-to-Weight Ratio'
# Assuming that 'Horsepower' and 'Number of Seating' are both numeric columns
data_df['Power-to-Weight Ratio'] = data_df['Horsepower'] / data_df['Number of Seating']

# 4. Create a binary feature for 'Has_Torque'
# Indicates whether the 'Torque' information is available or not
data_df['Has_Torque'] = data_df['Torque'].notnull()

# 5. Extract the year, month, and day from the 'Year' column
data_df['Year'] = pd.to_datetime(data_df['Year'], format='%Y')
data_df['Year'] = data_df['Year'].dt.year
data_df['Month'] = pd.to_datetime(data_df['Year'], format='%Y')
data_df['Month'] = data_df['Month'].dt.month
data_df['Day'] = pd.to_datetime(data_df['Year'], format='%Y')
data_df['Day'] = data_df['Day'].dt.day

# 6. Create dummy variables for categorical columns
# This step is necessary for some machine learning algorithms that cannot handle categorical data directly
categorical_columns = ['Company', 'Country of Origin', 'Transmission Type', 'Drivetrain', 'Looks', 'Body Type', 'Engine Type', 'Number of Cylinders', 'Brand', 'Price Category']
data_df = pd.get_dummies(data_df, columns=categorical_columns, drop_first=True)

# 7. Drop irrelevant columns
# Drop the original 'Model' column as we have extracted the 'Brand' from it
# Drop the 'Torque' column since we have created a binary 'Has_Torque' feature
data_df.drop(['Model', 'Torque'], axis=1, inplace=True)

# Display the updated summary of the data after data cleaning and feature engineering
print("Data Summary After Data Cleaning and Feature Engineering:")
print(data_df.info())

# Create a correlation matrix
correlation_matrix = data_df.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import re
import matplotlib.pyplot as plt

# Replace 'bikes_data.csv' with the actual file path if it's located in a different directory
file_path = 'bikes_data.csv'

# Read the CSV file into a pandas DataFrame
data_df = pd.read_csv(file_path)

# Data Cleaning and Feature Engineering Steps (same as before)

# ... (Refer to the previous code for data cleaning and feature engineering steps)

# Clean 'Horsepower' column by extracting numerical values and convert to numeric format (unit: 'hp')
data_df['Horsepower'] = data_df['Horsepower'].str.extract(r'(\d+\.?\d*)').astype(float)

# Clean 'Torque' column by extracting numerical values and convert to numeric format (unit: 'Nm')
data_df['Torque'] = data_df['Torque'].str.extract(r'(\d+\.?\d*)').astype(float)

# Handle missing values by imputing the mean of each column
data_df['Horsepower'].fillna(data_df['Horsepower'].mean(), inplace=True)
data_df['Torque'].fillna(data_df['Torque'].mean(), inplace=True)

# Select features for clustering (Horsepower and Torque)
features_for_clustering = ['Horsepower', 'Torque']
X_clustering = data_df[features_for_clustering]

# Standardize the features to have zero mean and unit variance
scaler = StandardScaler()
X_clustering_scaled = scaler.fit_transform(X_clustering)

# Determine the optimal number of clusters using the Elbow Method
inertia = []
for num_clusters in range(1, 11):
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(X_clustering_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Method to visualize the optimal number of clusters
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.grid(True)
plt.show()

# Based on the Elbow Method, we can choose the number of clusters (let's say 3 in this example)
num_clusters = 3

# Perform K-Means clustering with the chosen number of clusters
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(X_clustering_scaled)

# Add the cluster labels as a new column to the DataFrame
data_df['Cluster'] = cluster_labels

# Display the first few rows of the DataFrame with the cluster labels
print(data_df.head())

# Visualization of Clusters (Using Horsepower and Torque)
plt.figure(figsize=(10, 6))
for cluster_num in range(num_clusters):
    cluster_data = data_df[data_df['Cluster'] == cluster_num]
    plt.scatter(cluster_data['Horsepower'], cluster_data['Torque'], label=f'Cluster {cluster_num}', alpha=0.7)
plt.xlabel('Horsepower')
plt.ylabel('Torque')
plt.title('Clustering Based on Horsepower and Torque')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler

# Replace 'bikes_data.csv' with the actual file path if it's located in a different directory
file_path = 'bikes_data.csv'

# Read the CSV file into a pandas DataFrame
data_df = pd.read_csv(file_path)

# Data Cleaning and Feature Engineering Steps
# Remove non-numeric characters from the 'Horsepower' column and convert to numeric
data_df['Horsepower'] = data_df['Horsepower'].str.extract(r'(\d+\.?\d*)').astype(float)

# Remove non-numeric characters from the 'Torque' column and convert to numeric
data_df['Torque'] = data_df['Torque'].str.extract(r'(\d+\.?\d*)').astype(float)

# Drop rows where 'Horsepower' or 'Torque' is NaN after the extraction
data_df = data_df.dropna(subset=['Horsepower', 'Torque'])

# ... (Continue with the rest of the data cleaning and feature engineering steps)

# Select features for recommendation (Horsepower, Torque, and Number of Seating)
features_for_recommendation = ['Horsepower', 'Torque', 'Number of Seating']
X_recommendation = data_df[features_for_recommendation]

# Standardize the features to have zero mean and unit variance
scaler = StandardScaler()
X_recommendation_scaled = scaler.fit_transform(X_recommendation)

# Compute the cosine similarity matrix
cosine_sim = cosine_similarity(X_recommendation_scaled)


# Function to get bike recommendations based on similar features
def get_recommendations(bike_model, num_recommendations=5):
    try:
        idx = data_df[data_df['Model'] == bike_model].index[0]
    except IndexError:
        print(f"Bike model '{bike_model}' not found in the dataset.")
        return []

    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:num_recommendations + 1]
    bike_indices = [i[0] for i in sim_scores]
    return data_df['Model'].iloc[bike_indices].tolist()  # Convert to list

# Example usage of the recommendation system
bike_model = 'Shiver 900'  # Replace with the desired bike model name
num_recommendations = 5
recommended_bikes = get_recommendations(bike_model, num_recommendations)

if recommended_bikes:
    print(f"Recommended bikes for '{bike_model}':")
    print(recommended_bikes)
else:
    print(f"No recommendations found for '{bike_model}'.")

```
